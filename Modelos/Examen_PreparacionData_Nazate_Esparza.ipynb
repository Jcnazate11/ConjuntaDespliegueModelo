```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('data_evaluacion.csv')
data .head()
```

NOMBRAMIENTO DE NUEVAS ETIQUETAS

```python
# Nuevas etiquetas para las columnas
nuevas_columnas = ['edad', 'claseTrabajo', 'fnlwgt', 'educacion', 'educacionNum', 'estadoCivil','ocupacion', 'relacion', 
                   'raza', 'genero', 'gananciaCapital', 'perdidaCapital', 'horasSemana', 'pais', 'salario']

# Asignar nuevas etiquetas
data.columns = nuevas_columnas

```

```python
# Restablece el Ã­ndice
data.reset_index(drop=True, inplace=True)
```

LIMIPIEZA DATA

```python
# Contar los valores '?' en cada columna del DataFrame original
question_marks_count = (data == '?').sum()

# Mostrar el conteo de signos de interrogaciÃ³n en cada columna
print(question_marks_count)
```

```python

# Reemplazar los valores '?' por NaN
data.replace('?', pd.NA, inplace=True)
```

```python
categorical_columns = ['claseTrabajo', 'educacion', 'estadoCivil', 'ocupacion', 'relacion', 'raza', 'genero', 'pais']
for col in categorical_columns:
    mode_value = data[col].mode()[0]
    data[col] = data[col].fillna(mode_value)
```

```python
numerical_columns = ['edad', 'fnlwgt', 'educacionNum', 'gananciaCapital', 'perdidaCapital', 'horasSemana']
for col in numerical_columns:
    median_value = data[col].median()
    data[col] = data[col].fillna(median_value)
```

```python
data.head()
```

```python
data.info()
```

```python
# Elimina los espacios en blanco de los valores en la columna 'salary'
data['salario'] = data['salario'].str.strip()
salary_map = {'<=50K': 1, '>50K': 0}
data['salario'] = data['salario'].map(salary_map)
```

```python
print(data.head())
```

```python
data['genero'] = data['genero'].str.strip()
# Mapea los valores
data['genero'] = data['genero'].map({'Male': 1, 'Female': 0})
print(data.head())
```

```python
#Remplazamos estados civiles a Single y Couple
data['estadoCivil'] = data['estadoCivil'].str.strip()
data['estadoCivil'] = data['estadoCivil'].replace(['Divorced','Married-spouse-absent','Never-married','Separated','Widowed'],'Single')
data['estadoCivil'] = data['estadoCivil'].replace(['Married-AF-spouse','Married-civ-spouse'],'Couple')
data.head(10)

```

```python
data['pais'] = data['pais'].str.strip()

# Reemplazar valores en la columna 'pais'
data.loc[data['pais'] != 'United-States', 'pais'] = 'Non-US'
data.loc[data['pais'] == 'United-States', 'pais'] = 'US'
```

```python
data['pais'] = data['pais'].map({'US':1,'Non-US':0}).astype(int)
data.head(10)
```

```python
data['estadoCivil'] = data['estadoCivil'].map({'Couple':0,'Single':1}) 
data.head(10)
```

```python
label_encoder = LabelEncoder()
data['relacion'] = label_encoder.fit_transform(data['relacion'])
data.head(10)
```

```python
categories = label_encoder.classes_

# Crear un DataFrame para mostrar el mapeo
mapping_df = pd.DataFrame({
    'CategorÃ­a': categories,
    'CÃ³digo': range(len(categories))
})

print(mapping_df)
```

```python
def definir_tipo_empleo(x):
    if x['claseTrabajo'] == 'Federal-gov' or x['claseTrabajo']== 'Local-gov' or x['claseTrabajo']=='State-gov': return 'govt'
    elif x['claseTrabajo'] == 'Private':return 'private'
    elif x['claseTrabajo'] == 'Self-emp-inc' or x['claseTrabajo'] == 'Self-emp-not-inc': return 'self_employed'
    else: return 'without_pay'
 
data['claseTrabajo'] = data['claseTrabajo'].str.strip()  
data['tipoEmpleo']=data.apply(definir_tipo_empleo, axis=1)

data.head(10)
```

```python
label_encoder = LabelEncoder()
data['tipoEmpleo'] = label_encoder.fit_transform(data['tipoEmpleo'])
data.head(10)
```

```python
data['ocupacion'] = data['ocupacion'].replace(0, 'vacio')
```
