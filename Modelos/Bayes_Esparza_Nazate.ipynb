```python
import pandas as pd #manejo de archivos y analisis de datos
import numpy as np# dividir los datos yvectore matrices
import matplotlib.pyplot as plt # generar graficos histograficos 
import seaborn as sns

# operaciones de los modelos supervisados, modelar, entrenar
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc
from sklearn.preprocessing import StandardScaler
```

```python
data= pd.read_csv('new_data_evaluacion.csv')
data.head()
```

```python
data.describe()
```

```python
print(data.isnull().sum())
```

```python
data.drop("Unnamed: 0",axis=1,inplace=True)
```

```python
data.head()
```

```python
X = data.drop(['salario'], axis=1)
y = data['salario']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
# Crear el escalador y ajustarlo solo con el conjunto de entrenamiento
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
lrm =  GaussianNB()
lrm.fit(X_train,y_train)
```

```python
y_pred = lrm.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
confusion_matrix(y_test, y_pred)
```

```python
# Matriz de confusión
cm = confusion_matrix(y_test, y_pred)
```

```python
# Graficar la matriz de confusión
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Salario Bajo', 'Salario Alto'], yticklabels=['Salario Bajo', 'Salario Alto'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Matriz de Confusión para Naive Bayes')
plt.show()
```
